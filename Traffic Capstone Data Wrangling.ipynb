```python
import os
import pandas as pd

df= pd.read_csv('Metro_Interstate_Traffic_Volume.csv')
df.rain_1h.rename('rain')
df2 = df.rename(index=str, columns={"rain_1h":"Rain", "snow_1h":"Snow"})
def kelv_to_fahr(temp_kelv):
    temp_fahr = ((temp_kelv - 273.15) * (9/5)) + 32
    return temp_fahr

df2["temp"] = kelv_to_fahr(df2["temp"])
print(df2)
```

```python
df3= df2.set_index('holiday').groupby('holiday').mean()
print(df3.traffic_volume)
holiday_data= df3.traffic_volume
%store holiday_data
```

```python
df100= df2.set_index('temp').groupby('temp').mean()
print(df100.traffic_volume)
temp_data= df100.traffic_volume
%store temp_data
```
